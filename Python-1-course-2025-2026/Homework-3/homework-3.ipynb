```python
import json
import csv

# Задание 1 - словарь purchases
purchases = {}
with open('purchase_log.txt', 'r', encoding='utf-8') as f:
    next(f)
    for line in f:
        data = json.loads(line.strip())
        purchases[data['user_id']] = data['category']

print("Первые два элемента словаря purchases:")
for i, (user, cat) in enumerate(purchases.items()):
    print(user, cat)
    if i == 1:
        break


# Задание 2 - построчная обработка visit_log.csv
with open('visit_log.csv', 'r', encoding='latin-1') as f_in, \
     open('funnel.csv', 'w', encoding='utf-8', newline='') as f_out:
    
    reader = csv.DictReader(f_in)
    writer = csv.writer(f_out)
    
    writer.writerow(['user_id', 'source', 'category'])
    
    for row in reader:
        user_id = row.get('user_id')
        source = row.get('source')
        if user_id in purchases:
            writer.writerow([user_id, source, purchases[user_id]])

# Проверка: выводим первые три строки funnel.csv
print("\nПервые три строки funnel.csv:")
with open('funnel.csv', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        print(line.strip())
        if i == 2:
            break




```
