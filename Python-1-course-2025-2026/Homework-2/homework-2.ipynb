```python
# Задание 1
def sum_distance(from_, to_):
    if from_ > to_:
        from_, to_ = to_, from_
    return sum(range(from_, to_ + 1))

print(sum_distance(1, 5))   # 15 (1+2+3+4+5)
print(sum_distance(6, 1))   # 21 автоматически поменяет местами
print(sum_distance(4, 4))   # 4

# Задание 2
def trim_and_repeat(text, offset=0, repetitions=1):
    trimmed = text[offset:]
    return trimmed * repetitions

print(trim_and_repeat("Hello", 2, 3))   # "llollollo"
print(trim_and_repeat("Hello", 0, 2))   # "HelloHello"
print(trim_and_repeat("World", 3))      # "ld"
```
